### **tb_proyecto**


```sql
CREATE TABLE proyectos (
    id INT AUTO_INCREMENT PRIMARY KEY,
    nombre VARCHAR(255) NOT NULL
);
```

````python
lst_col_names = df_12.columns.values.tolist()
lista_proyectos = [col for col in lst_col_names if 'Proyecto' in col]
````

````python
import pyodbc

conn = pyodbc.connect(
    "DRIVER={MySQL ODBC 9.3 Unicode Driver};"
    "SERVER=localhost;"               # o la IP del servidor MySQL
    "DATABASE=gr4;"
    "UID=;"
    "PWD=;"
    "PORT=3306;"                      # asegúrate que sea el correcto
)
cursor = conn.cursor()

# Preparar la lista como tuplas
valores = [(nombre,) for nombre in lista_proyectos]

# Sentencia SQL parametrizada
sql = "INSERT INTO proyectos (nombre) VALUES (?)"

# simular la sentencia SQL
for record in lista_proyectos:
    print(sql.replace("?", "'{}'").format(record))
````

````python

# Ejecutar múltiples inserts
cursor.executemany(sql, valores)

# Confirmar cambios
conn.commit()
# conn.close()
````
